```python
# Q-8 Batsman record season wise
# Make a function which takes a input batsman_name and it returns a dataframe.
# Columns of the data frame are -
#  ['Season','Innings', 'TotalRuns', 'Avg', 'HighestScore','StrikeRate'].
# n result make Season column as index.

# Avg - total_runs/ no of time got out. - player_out column will help.

# StrikeRate -(total_runs/ balls faced) * 100- wides are not included in batsman ball faced counts.
# No balls are included. -> Extra_type column will help

# Batsman Can score runs on No Balls.

# Batsman can get out on No Ball or Wides. And even while being on non-striker. Keep these things
# in mind before masking.
import pandas as pd
stuff1=balls.merge(matches,on='ID')
stuff1['validbatsmanballs']=stuff1['extra_type'].apply(lambda x:1 if x!='wides' else 0)
def batsman_name(batsman):
  ss=stuff1[stuff1['batter']==batsman].copy()
  ss['no_of_outs']=ss['batter']==ss['player_out']
  ss1=ss.groupby(['Season','ID'],as_index=False)[['batsman_run','validbatsmanballs','no_of_outs']].sum()
  ss2=ss1.groupby('Season').agg({
      'ID':'count',
      'batsman_run':['sum','max'],
      'validbatsmanballs':'sum',
      'no_of_outs':'sum'
  })
  ss2['Avg']=(ss2[('batsman_run','sum')])/(ss2[('no_of_outs','sum')])
  ss2['innings']=ss2[('ID','count')]
  ss2['Total_runs']=ss2[('batsman_run','sum')]
  ss2['Highest_score']=ss2[('batsman_run','max')]
  ss2['balls_faced']=ss2[('validbatsmanballs','sum')]
  ss2['strike_rate']=(ss2['Total_runs']/ss2['balls_faced'])*100

  return ss2[['innings','Total_runs','Highest_score','balls_faced','strike_rate','Avg','no_of_outs']]
batsman_name('V Kohli')


```

```python
# Q-8 Batsman record season wise
# Make a function which takes a input batsman_name and it returns a dataframe. Columns of the data frame are -
#  ['Season','Innings', 'TotalRuns', 'Avg', 'HighestScore','StrikeRate'].

# In result make Season column as index.

# Avg - total_runs/ no of time got out. - player_out column will help.

# StrikeRate -(total_runs/ balls faced) * 100- wides are not included in batsman ball faced counts.
# No balls are included. -> Extra_type column will help

# Batsman Can score runs on No Balls.

# Batsman can get out on No Ball or Wides. And even while being on non-striker. Keep these things in mind before masking.
import pandas as pd
stuff1=balls.merge(matches,on='ID')
stuff1['validballs']=stuff1['extra_type'].apply(lambda x:1 if x!='wides' else 0)
def batsman_name(batsman):
  ss1=stuff1[stuff1['batter']==batsman]
  ss1['is_out']=ss1['batter']==ss1['player_out']
  ss2=ss1.groupby(['Season','ID'],as_index=False)[['batsman_run','validballs','is_out']].sum()
  ss3=ss2.groupby('Season').agg({
      'ID':'count',
      'batsman_run':['sum','max'],
      'validballs':'sum',
      'is_out':'sum'
  })
  ss3['Innings']=ss3[('ID','count')]
  ss3['Total_runs']=ss3[('batsman_run','sum')]
  ss3['Highest_score']=ss3[('batsman_run','max')]
  ss3['Average']=ss3['Total_runs']/ss3[('is_out','sum')]
  ss3['Strike_rate']=(ss3['Total_runs'])/(ss3[('validballs','sum')])*100
  return ss3[['Innings','Total_runs','Average','Highest_score','Strike_rate']]
batsman_name('MS Dhoni')
```
